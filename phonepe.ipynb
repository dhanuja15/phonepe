```python
import os
import json
import pandas as pd
import mysql.connector as db
```

```python
#AGGREGATED TRANSACTION DATA

agtr_path="C:/Users/lenovo/Desktop/New folder/phonepe pulse/pulse/data/aggregated/transaction/country/india/state/"
agtr_state_list=os.listdir(agtr_path) 

col={"state":[],"year":[],"quarter":[],"transaction_type":[],"transaction_count":[],"transaction_amount":[]}

for state in agtr_state_list:
    agtr_st_path=agtr_path+state+"/"
    agtr_yr_list=os.listdir(agtr_st_path)
    for year in agtr_yr_list:
        agtr_yr_path=agtr_st_path+year+"/"
        agtr_json_list=os.listdir(agtr_yr_path)
        for Jsonfile in agtr_json_list:
            agtr_json_path=agtr_yr_path+Jsonfile
            data=open(agtr_json_path,"r")
            d=json.load(data)
            for datas in d['data']['transactionData']:
                name=datas["name"]
                count=datas["paymentInstruments"][0]["count"]
                amount=datas["paymentInstruments"][0]['amount']
                col["transaction_type"].append(name)
                col["transaction_count"].append(count)
                col["transaction_amount"].append(amount)
                col["state"].append(state)
                col["year"].append(year)
                col["quarter"].append(int(Jsonfile.strip('.json')))

    
agg_trans_df=pd.DataFrame(col)  

agg_trans_df['state']=agg_trans_df['state'].str.replace('andaman-&-nicobar-islands','Andaman & Nicobar')
agg_trans_df['state']=agg_trans_df['state'].str.replace('dadra-&-nagar-haveli-&-daman-&-diu','Dadra and Nagar Haveli and Daman and Diu' )
agg_trans_df['state']=agg_trans_df['state'].str.replace('-'," ")
agg_trans_df['state']=agg_trans_df['state'].str.title()
            
```

```python
#AGGREGATED USER DATA

agus_path="C:/Users/lenovo/Desktop/New folder/phonepe pulse/pulse/data/aggregated/user/country/india/state/"
agus_state_list=os.listdir(agus_path)

col2={"state":[],"year":[],"quarter":[],"user_brand":[],"user_count":[],"user_percentage":[]}

for state in agus_state_list:
    agus_st_path=agus_path+state+"/"
    agus_yr_list=os.listdir(agus_st_path)
    for year in agtr_yr_list:
        agus_yr_path=agus_st_path+year+"/"
        agus_json_list=os.listdir(agus_yr_path)
        for Jsonfile in agus_json_list:
            agus_json_path=agus_yr_path+Jsonfile
            data=open(agus_json_path,"r")
            u=json.load(data)
            try:
                for i in u['data']['usersByDevice']:
                    brand=i["brand"]
                    count=i["count"]
                    percentage=i['percentage']
                    col2["user_brand"].append(brand)
                    col2["user_count"].append(count)
                    col2["user_percentage"].append(percentage)
                    col2["state"].append(state)
                    col2["year"].append(year)
                    col2["quarter"].append(int(Jsonfile.strip('.json')))
            except:
                pass

agg_user_df=pd.DataFrame(col2)

agg_user_df['state']=agg_user_df['state'].str.replace('andaman-&-nicobar-islands','Andaman & Nicobar')
agg_user_df['state']=agg_user_df['state'].str.replace('dadra-&-nagar-haveli-&-daman-&-diu','Dadra and Nagar Haveli and Daman and Diu' )
agg_user_df['state']=agg_user_df['state'].str.replace('-'," ")
agg_user_df['state']=agg_user_df['state'].str.title()
```

```python
#MAP USER DATA

map_us_path="C:/Users/lenovo/Desktop/New folder/phonepe pulse/pulse/data/map/user/hover/country/india/state/"
mapus_state_list=os.listdir(map_us_path)

col4={"state":[],"year":[],"quarter":[],"states/districts":[],"registeredUsers":[],"appOpens":[]}

for state in mapus_state_list:
    mapus_st_path=map_us_path+state+"/"
    mapus_yr_list=os.listdir(mapus_st_path)
    for year in mapus_yr_list:
        mapus_yr_path=mapus_st_path+year+"/"
        mapus_json_list=os.listdir(mapus_yr_path)
        for Jsonfile in mapus_json_list:
            mapus_json_path=mapus_yr_path+Jsonfile
            data=open(mapus_json_path,"r")
            m_u=json.load(data)
            
            for i in m_u['data']['hoverData'].items():
                districts=i[0]
                registeredusers=i[1]['registeredUsers']
                appopens=i[1]['appOpens']
                col4["states/districts"].append(districts)
                col4["registeredUsers"].append(registeredusers)
                col4["appOpens"].append(appopens)
                col4["state"].append(state)
                col4["year"].append(year)
                col4["quarter"].append(int(Jsonfile.strip('.json')))
           

map_user_df=pd.DataFrame(col4)

map_user_df['state']=map_user_df['state'].str.replace('andaman-&-nicobar-islands','Andaman & Nicobar')
map_user_df['state']=map_user_df['state'].str.replace('dadra-&-nagar-haveli-&-daman-&-diu','Dadra and Nagar Haveli and Daman and Diu' )
map_user_df['state']=map_user_df['state'].str.replace('-'," ")
map_user_df['state']=map_user_df['state'].str.title()
```

```python
#TOP TRANSACTION DATA

top_tr_path="C:/Users/lenovo/Desktop/New folder/phonepe pulse/pulse/data/top/transaction/country/india/state/"
toptr_state_list=os.listdir(top_tr_path)

col5={"state":[],"year":[],"quarter":[],"districts":[],"district_transaction_count":[],"district_transaction_amount":[]}
col6={"state":[],"year":[],"quarter":[],"pincodes":[],"pincode_transaction_count":[],"pincode_transaction_amount":[]}
      

for state in toptr_state_list:
    toptr_st_path=top_tr_path+state+"/"
    toptr_yr_list=os.listdir(toptr_st_path)
    for year in toptr_yr_list:
        toptr_yr_path=toptr_st_path+year+"/"
        toptr_json_list=os.listdir(toptr_yr_path)
        for Jsonfile in toptr_json_list:
            toptr_json_path=toptr_yr_path+Jsonfile
            data=open(toptr_json_path,"r")
            t_tr=json.load(data)
            try:
                for i in t_tr['data']['districts']:
                    districts=i['entityName']
                    di_count=i["metric"]["count"]
                    di_amount=i["metric"]['amount']
                    col5["districts"].append(districts)
                    col5["district_transaction_count"].append(di_count)
                    col5["district_transaction_amount"].append(di_amount)
                    col5["state"].append(state)
                    col5["year"].append(year)
                    col5["quarter"].append(int(Jsonfile.strip('.json')))
                for i in t_tr['data']['pincodes']:
                    pincodes=i['entityName']
                    pi_count=i["metric"]["count"]
                    pi_amount=i["metric"]['amount']
                    col6["pincodes"].append(pincodes)
                    col6["pincode_transaction_count"].append(pi_count)
                    col6["pincode_transaction_amount"].append(pi_amount) 
                    col6["state"].append(state)
                    col6["year"].append(year)
                    col6["quarter"].append(int(Jsonfile.strip('.json')))    
            except:
                pass      

top_transdis_df=pd.DataFrame(col5)
top_transpin_df=pd.DataFrame(col6)

top_transdis_df['state']=top_transdis_df['state'].str.replace('andaman-&-nicobar-islands','Andaman & Nicobar')
top_transdis_df['state']=top_transdis_df['state'].str.replace('dadra-&-nagar-haveli-&-daman-&-diu','Dadra and Nagar Haveli and Daman and Diu' )
top_transdis_df['state']=top_transdis_df['state'].str.replace('-'," ")
top_transdis_df['state']=top_transdis_df['state'].str.title()

top_transpin_df['state']=top_transpin_df['state'].str.replace('andaman-&-nicobar-islands','Andaman & Nicobar')
top_transpin_df['state']=top_transpin_df['state'].str.replace('dadra-&-nagar-haveli-&-daman-&-diu','Dadra and Nagar Haveli and Daman and Diu' )
top_transpin_df['state']=top_transpin_df['state'].str.replace('-'," ")
top_transpin_df['state']=top_transpin_df['state'].str.title()
```

```python
#Table creation 
mydb=db.connect(host="localhost",
                user='root',
                password='root',
                database='phonepe')
cursor=mydb.cursor()
```

```python
#Aggregated Transaction Table

query1='''create table if not exists aggregated_transaction (State varchar(255),
                                                                Year int, 
                                                                Quarter int, 
                                                                Transaction_Type varchar(255), 
                                                                Transaction_Count bigint,
                                                                Transaction_Amount bigint)'''
cursor.execute(query1)

sql='''insert into aggregated_transaction(State,Year, Quarter, Transaction_Type, Transaction_Count,
                                            Transaction_Amount)
                                            
                                            values(%s,%s,%s,%s,%s,%s)'''
val=agg_trans_df.values.tolist()
cursor.executemany(sql,val)
mydb.commit()
```

```python
#Aggregted User Table

cursor.execute('''create table if not exists aggregated_user (State varchar(255),
                                                                Year int, 
                                                                Quarter int, 
                                                                Brand varchar(255), 
                                                                Count bigint,
                                                                Percentage float)''')

sql='''insert into aggregated_user(State,Year, Quarter,Brand, Count, Percentage)
                                            
                                            values(%s,%s,%s,%s,%s,%s)'''
val=agg_user_df.values.tolist()
cursor.executemany(sql,val)
mydb.commit()
```
